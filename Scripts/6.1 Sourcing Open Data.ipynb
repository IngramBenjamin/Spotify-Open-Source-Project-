# 04. Data Consistency Checks

```python
# Creating new dataframe of duplicate rows 
df_dups = df[df.duplicated()]
```

```python
df_dups.shape
```

```python
# Removing duplicate rows from original dataframe 
df_no_dups = df.drop_duplicates()
```

```python
df_no_dups.shape
```

```python
# Removing erroneous row with column headers as values
df_no_dups.drop(df_no_dups[df_no_dups['rank'] == 'rank'].index, inplace = True)
```

```python
df_no_dups.shape
```

```python
# Checking for missing values 

df_no_dups.isnull().sum()
```

```python
# Creating a dataframe of the missing values 

df_nan = df_no_dups[df_no_dups['key'].isnull() == True]
```

```python
df_nan.head(50)
```

```python
df_no_dups.dropna(inplace = True)
```

```python
df_no_dups.shape
```

```python
# Checking for mixed datatypes

for col in df_no_dups.columns.tolist():
  weird = (df_no_dups[[col]].applymap(type) != df_no_dups[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_no_dups[weird]) > 0:
    print (col)
```

```python
# Changing Data Types to proper numeric type

df_no_dups['rank'] = df_no_dups['rank'].astype('int')
df_no_dups['number_of_artists'] = df_no_dups['number_of_artists'].astype('int')
df_no_dups['collab'] = df_no_dups['collab'].astype('bool')
df_no_dups['number_of_album_tracks'] = df_no_dups['number_of_album_tracks'].astype('int')
df_no_dups['peak_rank'] = df_no_dups['peak_rank'].astype('int')
df_no_dups['previous_rank'] = df_no_dups['previous_rank'].astype('int')
df_no_dups['weeks_on_chart'] = df_no_dups['weeks_on_chart'].astype('int')
df_no_dups['weekly_streams'] = df_no_dups['weekly_streams'].astype('int')
df_no_dups['danceability'] = df_no_dups['danceability'].astype('float')
df_no_dups['energy'] = df_no_dups['energy'].astype('float')
df_no_dups['key'] = df_no_dups['key'].astype('int')
df_no_dups['mode'] = df_no_dups['mode'].astype('int')
df_no_dups['loudness'] = df_no_dups['loudness'].astype('float')
df_no_dups['speechiness'] = df_no_dups['speechiness'].astype('float')
df_no_dups['acousticness'] = df_no_dups['acousticness'].astype('float')
df_no_dups['instrumentalness'] = df_no_dups['instrumentalness'].astype('float')
df_no_dups['liveness'] = df_no_dups['liveness'].astype('float')
df_no_dups['valence'] = df_no_dups['valence'].astype('float')
df_no_dups['tempo'] = df_no_dups['tempo'].astype('float')
df_no_dups['duration'] = df_no_dups['duration'].astype('int')
```

```python
df_no_dups.head()
```

```python
# Confirming amended data types 
df_no_dups.info()
```

```python
# Recalculating descriptive statistics after consistency checks 
pd.set_option('display.float_format', lambda x: '%.2f' % x)
df_no_dups.describe()
```

```python
# Updating -1 value in 'previous_rank' column to 'n/a' after review found that this likely indicated songs that
# were not charted in the previous week
df_no_dups['previous_rank'] = df_no_dups['previous_rank'].replace(-1,'N/A')
```

```python
# Checking values for 'previous_rank' column to confirm value change
df_no_dups['previous_rank'].value_counts(dropna = False)
```

```python
df_no_dups['release_date'] = df_no_dups['release_date'].replace('1967-09','09/01/1967')
df_no_dups['release_date'] = df_no_dups['release_date'].replace('1957-09','09/01/1957')
```
